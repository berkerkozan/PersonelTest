```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'foo' : ['one','one','one','two','two','two'],
     'bar' : ['A','B','C','A','B','C'],
                  'baz':[1,2,3,4,5,6]
                  })
df
```

```python
df.pivot('foo', 'bar')
```

```python
print pd.pivot_table(df,index=["bar"])
df.groupby("bar").sum()
```

```python
columns = pd.MultiIndex.from_tuples([('A', 'cat'), ('B', 'dog'),
           ('B', 'cat'), ('A', 'dog')],
          names=['exp', 'animal'])
index = pd.MultiIndex.from_product([('bar', 'baz', 'foo', 'qux'),
                         ('one', 'two')],
                                   
                        names=['first', 'second'])
df = pd.DataFrame(np.random.randn(8, 4), index=index, columns=columns)
df
```

```python
df.unstack()
```
